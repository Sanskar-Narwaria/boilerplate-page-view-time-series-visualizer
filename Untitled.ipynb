```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("fcc-forum-pageviews.csv",parse_dates=['date'])
```

```python
df
```

```python
lower_bound = df['value'].quantile(0.025)
upper_bound = df['value'].quantile(0.975)

# Filter out the rows outside the specified bounds
df = df[(df['value'] >= lower_bound) & (df['value'] <=upper_bound)]
```

```python
df
```

```python
df_filtered=df
```

```python
fig,ax=plt.subplots(figsize=(12,10))
ax.plot(df_filtered['date'],df_filtered['value'],color='red',linewidth=1)
ax.set_xlabel('Date')
ax.set_ylabel('Page Views')
ax.set_title('Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
```

```python
df_filtered
```

```python
df_filtered['month']=df_filtered['date'].dt.month_name()
df_filtered['day']=df_filtered['date'].dt.day
df_filtered['year']=df_filtered['date'].dt.year
```

```python
df_filtered
```

```python
df=df_filtered.copy()
```

```python
grouped_df1=df_filtered.groupby(['year','month'])
```

```python
new_df=pd.DataFrame()
```

```python
new_df['mean']=grouped_df1['value'].mean()
```

```python
new_df
```

```python
newDf2=new_df.reset_index()
```

```python
newDf2
```

```python
month_order = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}

# Convert month column to a categorical type with custom order
newDf2['month'] = pd.Categorical(newDf2['month'], categories=month_order.keys(), ordered=True)

# Sort the DataFrame by the categorical month column
df_sorted = newDf2.sort_values(by=['year','month'])
```

```python
df_sorted
```

```python
import matplotlib as mpl
fig, ax = plt.subplots(figsize=(10, 10))

# Create the bar plot
sns.barplot(data=df_sorted, x='year', y='mean', hue='month', ax=ax,palette='tab10')

ax.set_xlabel('Years')
ax.set_ylabel('Average Page Views')
ax.legend_.set_title("Months")
actual = len([rect for rect in ax.get_children() if isinstance(rect, mpl.patches.Rectangle)])
actual
```

```python
new_df=pd.DataFrame(grouped_df1['value'].sum()).reset_index()
```

```python
new_df
```

```python
df_filtered['month'] = pd.Categorical(df_filtered['month'], categories=month_order.keys(), ordered=True)

# Sort the DataFrame by the categorical month column
df_sorted = df_filtered.sort_values(by=['month'])
df_sorted
```

```python
box_df=df_sorted[['month','year','value']].rename(columns={'month':'Month','year':'Year','value':'Page Views'})
month_abbr = {
    'January': 'Jan', 'February': 'Feb', 'March': 'Mar', 'April': 'Apr', 'May': 'May', 'June': 'Jun',
    'July': 'Jul', 'August': 'Aug', 'September': 'Sep', 'October': 'Oct', 'November': 'Nov', 'December': 'Dec'
}

# Map the full month names to abbreviated forms
box_df['Month'] = box_df['Month'].map(month_abbr)
box_df
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))

sns.boxplot(data=box_df,x='Year',y='Page Views',ax=ax[0],hue='Year',legend=False,palette=['blue', 'orange','green','red'],flierprops={"marker": "d", "markersize": 2,"markerfacecolor": "black"})
ax[0].set_title('Year-wise Box Plot (Trend)')
ax[0].grid(False)
sns.boxplot(data=box_df,x='Month',y='Page Views',ax=ax[1],hue='Month',legend=False,showcaps=False,flierprops={"marker": "d", "markersize": 2,"markerfacecolor": "black"})
ax[1].set_title('Month-wise Box Plot (Seasonality)')
ax[1].grid(False)

```
